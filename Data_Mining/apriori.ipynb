```python
from efficient_apriori import apriori
```

```python
transactions = [['eggs', 'chicken', 'soup'],
                ['eggs', 'chicken', 'apple'],
                ['soup', 'chicken'],
                ['eggs', 'meat', 'banana'],
                ['bacon', 'meat', 'apple'],
                ]

itemsets, rules = apriori(transactions, min_support=0.2,  min_confidence=1)

for x in rules:
    print ("-------------------------------------")
    print ("Data : "+str(x.lhs))
    print ("Support : "+str(x.support))
    print ("Confidence : "+str(x.confidence))
    print ("Lift : "+str(x.lift))
    print ("Conviction : "+str(x.conviction))
    print ("-------------------------------------")
    
    
rules_rhs = filter(lambda rule: len(rule.lhs) == 2 and len(rule.rhs) == 1, rules)

for rule in sorted(rules_rhs, key=lambda rule: rule.lift):
  print(rule)
```
