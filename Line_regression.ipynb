```python
# Source :
# https://www.edureka.co/blog/linear-regression-in-python/

%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
# Reading Data
data = pd.read_csv('fuel_consumption.csv')

print(data.shape)
data.head()
```

```python
# Collecting X and Y
x = data["FUELCONSUMPTION_COMB"].values
y = data["CO2EMISSIONS"].values
```

```python
# Mean X and Y
mean_x = np.mean(x)
mean_y = np.mean(y)
```

```python
# Total number of values
n = len(x)
```

```python
# Using the formula to calculate m and c
numer = 0
denom = 0
for i in range(n):
    numer += (x[i] - mean_x) * (y[i] - mean_y)
    denom += (x[i] - mean_x) ** 2
    m = numer / denom
    print("gradient : "+str(m))
    c = mean_y - (m * mean_x)
    print("coeficient : "+str(c))
```

```python
# Plotting Values and Regression Line
max_x = np.max(x) + 100
min_x = np.min(x) - 100

# Calculating line values x and y
x = np.linspace(min_x, max_x, 50)
y = c + m * x 

print (y)
```

```python
# Ploting Line
ax = plt.plot(x, y, color='#52b920', label='Regression Line')
# Ploting Scatter Points
ax = plt.scatter(x, y, c='#ef4423', label='Scatter Plot')
plt.xlabel('FUEL CONSUMPTION')
plt.ylabel('CO2 EMISSIONS')
plt.legend()
plt.show()
```
